```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd

points = pd.read_csv('KeyboardInput.csv',sep=';')

fig = plt.figure()
ax = fig.add_subplot(111)

x = points['xPos'].values
y = points['yPos'].values
actions = points['input'].values

#ax.set_xlim(40, 60)  # Setze die gewünschten Grenzwerte für die x-Koordinate
#ax.set_ylim(40, 70)

ax.set_xlabel('X-Achse')
ax.set_ylabel('Y-Achse')

ax.plot(x, y, c='#0077B6', label="Trajektion")

for i, actions in enumerate(actions):
    if i == 0:
        continue
        
    if(actions != 'Unknown'):
        plt.plot( x[i], y[i], c="#EE6C4D", marker="x")
        plt.text( x[i] + 0.07, y[i] + 0.07, actions, fontsize=9)
        

# zeigt die Kreise an
#p + 0.1lt.scatter(x + 0.1, y, s=2000, edgecolor='black', facecolor='none')

#plot first and last point
#ax.plot(first_point[0], first_point[1], c='#EE6C4D', marker='x')
#ax.plot(last_point[0], last_point[1], c='#EE6C4D', marker='x')

plt.legend()

# Fügt ein Label zum Start- und Endpunkt hinzu.
#ax.text (first_point[0], first_point[1], 'Start', color='black') 
#ax.text (last_point[0], last_point[1], 'Ende', color='black') 

plt.show()
```

```python
csv_files = ['Datei1.csv', 'Datei2.csv', 'Datei3.csv', 'Datei4.csv', 'Datei5.csv',
             'Datei6.csv', 'Datei7.csv', 'Datei8.csv', 'Datei9.csv', 'Datei10.csv']

fig, ax = plt.subplots()

for csv_file in csv_files:
    # Lade die Daten aus der CSV-Datei
    df = pd.read_csv(csv_file, sep=';')

    x = df['xPos'].values
    y = df['yPos'].values

    x_transformed = (x - 50) * 10
    y_transformed = (y - 50) * 10

    ax.plot(x_transformed, y_transformed, marker='x', label=csv_file)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()

plt.show()
```

```python
points = pd.read_csv('TestFlug.csv',sep=';')
#points = pd.read_csv('BA_Ergebniss_Demo.csv',sep=';')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = points['xPos'].values
y = points['yPos'].values
z = points['zPos'].values

first_point = (x[0], y[0], z[0])
last_point = (x[-1], y[-1], z[-1])

# Plot all points
ax.plot(x, y, z, c='#0077B6', marker='x')
ax.scatter(x, y, z, s=200, edgecolor='black', facecolor='none')

#plot first and last point
ax.plot(first_point[0], first_point[1], first_point[2], c='#EE6C4D', marker='x')
ax.plot(last_point[0], last_point[1], last_point[2], c='#EE6C4D', marker='x')

# Fügt ein Label zum Start- und Endpunkt hinzu.
ax.text (first_point[0], first_point[1], first_point[2], 'Start', color='black') 
ax.text (last_point[0], last_point[1], last_point[2], 'Ende', color='black') 

#ax.set_xlim(30, 70)  # Setze die gewünschten Grenzwerte für die x-Koordinate
#ax.set_ylim(30, 70)
ax.set_zlim(0, 100)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.view_init(elev=20, azim=220) # Ändert den Betrachtungswinkel

plt.show()

```
