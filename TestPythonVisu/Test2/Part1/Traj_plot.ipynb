```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd

points = pd.read_csv('BA_Exp2_Trajectory.csv',sep=';')

fig = plt.figure()
ax = fig.add_subplot(111)

x = points['xPos'].values
y = points['yPos'].values
actions = points['input'].values

x_transformed = (x - 50) / 10
y_transformed = (y - 50) / 10

ax.set_xlabel('X in Meter')
ax.set_ylabel('Y in Meter')

ax.plot(x_transformed, y_transformed, c='#0077B6')

action_label_set = False;

for i, action in enumerate(actions):
    if i == 0:
        continue
        
    if(action != 'Unknown'):
        if not action_label_set:
            plt.plot( x_transformed[i], y_transformed[i], c="#0077B6", marker="x", mec="red", label="Befehl")
        elif action_label_set == True:
            plt.plot( x_transformed[i], y_transformed[i], c="#0077B6", marker="x", mec="red")
        plt.text( x_transformed[i] + 0.02, y_transformed[i] + 0.02, action, fontsize=9)
        action_label_set = True
 
ax.set_xlim(-0.1, 1.5)
ax.set_ylim(-0.2, 2)

ax.grid(True)

plt.legend()

plt.savefig('Exp2_DemoTraj.pdf', format='pdf')

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd
import os

csv_files = ['Wiederholungen/Exe1.csv', 'Wiederholungen/Exe2.csv', 'Wiederholungen/Exe3.csv', 
             'Wiederholungen/Exe4.csv', 'Wiederholungen/Exe5.csv', 'Wiederholungen/Exe6.csv', 
             'Wiederholungen/Exe7.csv', 'Wiederholungen/Exe8.csv', 'Wiederholungen/Exe9.csv', 
             'Wiederholungen/Exe10.csv', 'BA_Exp2_Trajectory.csv' ]

fig, ax = plt.subplots()
set_retry_label = False;

for csv_file in csv_files:
    # Lade die Daten aus der CSV-Datei
    df = pd.read_csv(csv_file, sep=';')
    
    x = df['xPos'].values
    y = df['yPos'].values

    x_transformed = (x - 50) / 10
    y_transformed = (y - 50) / 10

    if os.path.basename(csv_file) == 'BA_Exp2_Trajectory.csv':
        ax.plot(x_transformed, y_transformed, c="#0077B6", label="Soll-Trajektion")
    else:
        if not set_retry_label:
            ax.plot(x_transformed, y_transformed, c="orange", label="Ist-Trajektion")
            set_retry_label = True;
        else:
            ax.plot(x_transformed, y_transformed, c="orange")
        

ax.set_xlabel('X in Meter')
ax.set_ylabel('Y in Meter')
ax.legend();
ax.grid(True)

plt.savefig('Exp2_berlapptTraj.pdf', format='pdf')

plt.show()
```

```python
# Aktion 1 - TakeOff
curr_action = first_action
origin = curr_action[-1]
origin_x = (origin[2] - 50) / 10 # skaliere von cm in m und versetze die pos auf (0,0)
origin_y = (origin[3]  - 50) / 10
shift_x = origin_x
shift_y = origin_y

shifted_origin_x = origin_x - shift_x
shifted_origin_y = origin_y - shift_y

fig, ax = plt.subplots()

ax.axhline(y=shifted_origin_y, color='gray')  # Horizontale Linie durch den Ursprung
ax.axvline(x=shifted_origin_x, color='gray')  # Vertikale Linie durch den Ursprung

# Fge Achsentitel hinzu
ax.set_xlabel('X in Meter')
ax.set_ylabel('Y in Meter')

# Setze das Koordinatensystem mit angepassten Achsenbegrenzungen
#ax.set_xlim(-0.2, 0.2)
#ax.set_ylim(-0.2, 0.2)

ax.plot(shifted_origin_x, shifted_origin_y, marker='o', color='#0077B6', label='Soll-Koordinate')

label_set = False;

for i in range(len(curr_action) - 1):
   
    pointX = ((curr_action[i][2]  - 50) / 10) - shift_x
    pointY = ((curr_action[i][3]  - 50) / 10) - shift_y
    abweichungs_koordinaten.append((pointX, pointY))
    if not label_set:
        ax.plot(pointX, pointY, marker='o', color='orange', label='Ist-Koordinate')
        label_set = True
    else:
        ax.plot(pointX, pointY, marker='o', color='orange')
        
# Zeichne das Koordinatensystem

plt.savefig('Exp2_OhneTol_Act1.pdf', format='pdf')
ax.legend()
print(len(first_action))
plt.show()
```
