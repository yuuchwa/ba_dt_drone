```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd

points = pd.read_csv('Rechteck/Rechteck.csv',sep=';')

fig = plt.figure()
ax = fig.add_subplot(111)

x = points['xPos'].values
y = points['yPos'].values

x_transformed = (x - 50) / 10
y_transformed = (y - 50) / 10

first_point = (x_transformed[0], y_transformed[0])
last_point = (x_transformed[-1], y_transformed[-1])

ax.set_xlabel('X in Meter')
ax.set_ylabel('Y in Meter')

ax.grid(True)

ax.plot(x_transformed, y_transformed, c='#0077B6', marker='x')

#plot first and last point
ax.plot(first_point[0], first_point[1], c='#EE6C4D', marker='o')
ax.plot(last_point[0], last_point[1], c='#EE6C4D', marker='o')

# Fügt ein Label zum Start- und Endpunkt hinzu.
ax.text (first_point[0] + 0.03, first_point[1], 'TakeOff', color='black') 
ax.text (last_point[0] - 0.04, last_point[1] + 0.07, 'Land', color='black') 

plt.savefig('Exp1_RechteckTraj.pdf', format='pdf')

plt.show()
```
